```python
# merged two dataset on 'title' column
merged_movie_data = movie_df.merge(credit_df, on='title')
merged_movie_data.columns
```

```python
# make a fresh data with some selected columns
movie_data = merged_movie_data[['movie_id', 'title', 'overview','genres','keywords','cast','crew']]
movie_data.head(2)
```

```python
import ast
import warnings
warnings.filterwarnings("ignore")

def convert_dictTojson(text):
    """
    convert_dictTojson(text) converts a string text containing a JSON-like structure 
    into a list of names extracted from each dictionary in the structure. 
    """
    try:
        data = ast.literal_eval(text)  # Safely evaluate the string as a Python literal or container
        names = [item['name'] for item in data if isinstance(item, dict) and 'name' in item]
        return names
    except (SyntaxError, ValueError) as e:
        print(f"Error evaluating text: {e}")
        return []
```

```python
movie_data.dropna(inplace=True)
```

```python
movie_data['genres'] = movie_data['genres'].apply(convert_dictTojson)
movie_data.head(2)
```

```python
movie_data['keywords'] = movie_data['keywords'].apply(convert_dictTojson)
movie_data.head(2)
```

```python
movie_data['genres'].unique
```

```python
movie_data['cast'] = movie_data['cast'].apply(convert_dictTojson)
movie_data.head(2)
```

```python
movie_data['cast'] = movie_data['cast'].apply(lambda x:x[0:3])
movie_data['cast']
```

```python
def find_director_name(text):
    """
    find_director_name(text) function, 
    which extracts names of directors from a JSON-like structure stored in the string text,
    """
    try:
        data = ast.literal_eval(text)
        director_names = [item['name'] for item in data if isinstance(item, dict) and item.get('job') == 'Director']
        return director_names
    except (SyntaxError, ValueError) as e:
        return []
```

```python
movie_data['crew'] = movie_data['crew'].apply(find_director_name)
movie_data['crew']
```

```python
movie_data['cast'] 
```

```python
def remove_spaces_from_strings(List):
    """
    remove_spaces_from_strings(List) function takes a list "List" containing strings as input. 
    It iterates through each string in "List", removes any spaces within each string using .replace(" ", ""), 
    and then returns a new list "List1" with the modified strings.
    """
    return [i.replace(" ", "") for i in List] 
```

```python
movie_data['cast']     = movie_data['cast'].apply(remove_spaces_from_strings)
movie_data['crew']     = movie_data['crew'].apply(remove_spaces_from_strings)
movie_data['genres']   = movie_data['genres'].apply(remove_spaces_from_strings)
movie_data['keywords'] = movie_data['keywords'].apply(remove_spaces_from_strings)

movie_data.head(2)
```

```python
movie_data['overview'] = movie_data['overview'].apply(lambda x:x.split())
movie_data['tags'] = movie_data['overview'] + movie_data['genres'] + movie_data['keywords'] + movie_data['cast'] + movie_data['crew']
movie_data.head(2)
```

```python
new_movie_data = movie_data.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
new_movie_data['tags'] = new_movie_data['tags'].apply(lambda x: " ".join(x))
new_movie_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=3000, stop_words='english')
vector = cv.fit_transform(new_movie_data['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```
